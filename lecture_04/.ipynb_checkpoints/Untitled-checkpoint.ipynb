# Anwendungsbeispiel: Web-Crawler & Suchmaschine

```python
import requests 
import lxml.html as lh

base = 'http://www.fh-swf.de'
visited = set()
todo = { base }

while len(todo) > 0:
    url = todo.pop()
    print(f"crawling {url}")
    visited.add(url)
    resp = requests.get(url)
    doc = lh.fromstring(resp.content)
    for link in doc.xpath('//a'):
        href = link.get('href')
        if href is None or href[0] == '#' or href.startswith('https://') or href.startswith('http://'):
            continue
        if href[0] == '/':
            href = base + "/" + href
        else:
            href = url + "/" + href
        print(f"href: {href}")
        
```
