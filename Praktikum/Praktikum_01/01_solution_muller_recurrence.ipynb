```python
from decimal import Decimal
from fractions import Fraction

N = 80

def muller_float(y, z):
    """Naive Version mit float"""
    return 108 - (815 - 1500 / z) / y

def muller_decimal(y, z):
    """Version mit decimal"""
    return Decimal(108) - (Decimal(815) - Decimal(1500) / z) / y

def muller_fraction(y, z):
    """Version mit fraction"""
    return Fraction(108) - (Fraction(815) - Fraction(1500) / z) / y

def do_recurrence(E, x):
    for n in range(2, N+1):
        x.append(E(x[n-1], x[n-2]))

x_float = [ 4, 4.25 ]
do_recurrence(muller_float, x_float)

x_decimal = [ Decimal(4), Decimal(4.25) ]
do_recurrence(muller_decimal, x_decimal)

x_fraction = [ Fraction(4), Fraction(4.25) ]
do_recurrence(muller_fraction, x_fraction)


for n in range(N+1):
    print('{:2d} {:22.15f} {:22.15f} {:22.15f}'.format(n, x_float[n], x_decimal[n], float(x_fraction[n])))

```
