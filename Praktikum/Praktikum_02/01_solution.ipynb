## Vorbereitung
Die folgende Zelle erzeugt zunächst ein Array mit 1.000 zufälligen Zahlen und stellt diese mit Hilfe von `matplotlib` dar (diese Bibliothek werden wir in der Vorlesung noch genauer behandeln).

```python
# Wir benötigen Zufallszahlen und eine Bibliothek zum plotten
%matplotlib inline
import random
import matplotlib.pyplot as plt

N = 1000

# Eintausend Zufallszahlen zwischen 0 und 1
unsorted = [ random.random() for i in range(N)]

# Die folgenden Zeilen zeigen die Zahlen als Scatter-Plot an
plt.scatter(range(N), unsorted)
plt.show()
```

## Jetzt sind Sie an der Reihe!
Schreiben Sie eine Funktion `quicksort(lst)`, die eine übergebene Liste sortiert. Sie sollen dabei **nicht** auf die eingebauten Sortierfunktionen von Python zurückgreifen, sondern den Algorithmus selbst implementieren.

```python
# Bitte implementieren Sie den Algorithmus hier!
def quicksort(lst: list, links: int=0, rechts: int=N-1):
    """Diese Funktion sortiert die übergebene Liste ('in-place sort')
    
    Argumente:
    lst    -- zu sortierende Liste
    links  -- Startindex des zu sortierenden Bereich (inklusiv)
    rechts -- Endindex des zu sortierenden Bereichs (exklusiv)
    """
    
    if links < rechts:
        # Wähle ein zufälliges Element als Pivot
        pivot_idx = random.randint(links, rechts)
        pivot = lst[pivot_idx]
        #pivot = lst[rechts]
        
        i = links
        j = rechts
        
        while i < j:
            # Suche von links nach Element größer als das Pivotelement
            while (i < j) & (lst[i] < pivot):
                i += 1
                
            # Suche von rechts nach Element kleiner als das Pivotelement
            while (j > i) & (lst[j] > pivot):
                j -= 1
                
            # Tausche Elemente
            lst[i], lst[j] = lst[j], lst[i]
        
        # Plotte Zwischenergebnis zur Visualisierung
        # plt.scatter(range(N),lst)
        # plt.show()
        
        # Rekursion für die Teile links und rechts from Pivot
        quicksort(lst, links, j - 1)
        quicksort(lst, i + 1, rechts)
    
```

```python
quicksort(unsorted)
```

## Test des Ergebnisses
Der folgende Code plotted das Ergebnis und testet, ob die Liste tatsächlich sortiert ist.

```python
# Prüfe, ob die Liste sortiert ist
for i in range(N-1):
    if unsorted[i] > unsorted[i+1]:
        print("Fehler an Stelle {0:d}: {1:f} > {2:f} !".format(i, unsorted[i], unsorted[i+1]))
        break;
else:
    print("OK, die Liste ist sortiert!")

# Die folgenden Zeilen zeigen die Zahlen als Scatter-Plot an
plt.scatter(range(N), unsorted)
plt.show()
```
